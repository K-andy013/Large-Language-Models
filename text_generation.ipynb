```python
from groq import Groq #importing Groq Client object
client=Groq(api_key=apiKey)
```

```python
models=["mixtral-8x7b-32768","llama3-groq-70b-8192-tool-use-preview","llama-3.1-70b-versatile","gemma2-9b-it"]
```

```python
for model in models:
  chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Concisely summarize the triple quote enclosed text into a single paragraph. Use line breaks to make th etext easier to read\
            *** The enigmatic allure of the ancient city of Machu Picchu, nestled high in the Peruvian Andes,\
             continues to captivate explorers and historians alike. Hidden for centuries amidst mist-shrouded\
              mountains, this lost city of the Incas offers a glimpse into a bygone era of advanced civilization.\
               Its intricate stonework, terraced agriculture, and sophisticated astronomical alignments stand \
               as a testament to the ingenuity of the Inca people.Theories abound regarding the city's purpose,\
                ranging from a sacred religious site to a strategic military fortress. Some scholars speculate that\
                 Machu Picchu may have served as a royal retreat for the Inca elite, a place where they could escape\
                  the bustling urban centers and connect with the divine. Others propose that it was a center for\
                   astronomical observations and spiritual ceremonies, aligning with the Inca belief in the\
                    interconnectedness of the cosmos and the earthly realm.Regardless of its exact function,\
                     Machu Picchu remains a marvel of human achievement.Its remote location and challenging \
                     terrain have contributed to its preservation, allowing visitors to experience the awe-inspiring\
                      ruins firsthand. As we wander through the ancient streets and gaze upon the majestic peaks of\
                       the Andes, we can't help but marvel at the ingenuity and resilience of the people who once called\
                        this extraordinary place home.***",
        }
    ],
    model=model,
)
  print(model+"\n")
  print(chat_completion.choices[0].message.content+"\n")
```
